## Viewport

```openscad
cube([20,20,20]);
%display
```

Usually the object is displayed in a "comprehensive" way. You can adjust the viewport however with the following variables:

- __\\$vpr__: viewport rotation angles in degrees
- __\\$vpt__: viewport translation
- __\\$vpd__: viewport camera distance


```openscad
$vpt = [ 0,0,0 ];
$vpr = [ 30,0,20 ];
$vpd = 150;

cube([20,20,20]);
%display
```
